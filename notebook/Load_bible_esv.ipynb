```python

import xml.etree.ElementTree as ET
from google.colab import files
import csv
uploaded= files.upload()
# Load and parse the XML file
tree = ET.parse("ESV.xml")  # replace with your file path if needed
root = tree.getroot()

# Open CSV file to write
with open("Bible_ESV.csv", mode="w", newline='', encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Book", "Chapter", "Verse", "Text"])  # Header

    for book in root.findall("b"):
        book_name = book.get("n")
        for chapter in book.findall("c"):
            chapter_num = chapter.get("n")
            for verse in chapter.findall("v"):
                verse_num = verse.get("n")
                verse_text = verse.text.strip() if verse.text else ""
                writer.writerow([book_name, chapter_num, verse_num, verse_text])
```
